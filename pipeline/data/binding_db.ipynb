```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator
from Bio.SeqUtils import molecular_weight
from sklearn.cluster import KMeans
from scipy.spatial.distance import pdist, squareform
import swifter  # if installed for faster apply

# 1. Load the BindingDB TSV file
binding_data_path = "BindingDB_All.tsv"
# For demonstration, read the first 100,000 rows.
binding_data = pd.read_csv(binding_data_path, nrows=100_000, sep='\t', on_bad_lines='warn')
initial_total = len(binding_data)

# 2. Filter to only include rows with a valid IC50 measurement
binding_data["IC50 (nM)"] = pd.to_numeric(binding_data["IC50 (nM)"], errors='coerce')
filtered_data = binding_data.dropna(subset=["IC50 (nM)"]).copy()
print(f"IC50 Filtering: Removed {(initial_total - len(filtered_data)) / initial_total * 100:.2f}% of rows.")

# 3. Protein Filtering: Remove proteins with problematic sequences
# Determine the appropriate sequence column.
if "BindingDB Target Chain Sequence" in filtered_data.columns:
    seq_column = "BindingDB Target Chain Sequence"
elif "BindingDB Target ChainSequence" in filtered_data.columns:
    seq_column = "BindingDB Target ChainSequence"
else:
    seq_column = None

# Get unique proteins based on BindingDB MonomerID.
protein_groups = filtered_data.dropna(subset=["BindingDB MonomerID"]).drop_duplicates(subset=["BindingDB MonomerID"])
initial_protein_count = len(protein_groups)
valid_protein_rows = []

for idx, row in protein_groups.iterrows():
    seq = row[seq_column] if seq_column is not None else ""
    if not seq or pd.isna(seq):
        print(f"Skipping protein {row['BindingDB MonomerID']} due to missing sequence.")
        continue
    try:
        _ = molecular_weight(seq, seq_type='protein')
        valid_protein_rows.append(row)
    except Exception as e:
        print(f"Skipping protein {row['BindingDB MonomerID']} due to sequence error: {e}")
        
valid_protein_df = pd.DataFrame(valid_protein_rows)
valid_protein_count = len(valid_protein_df)
print(f"Protein Filtering: Removed {(initial_protein_count - valid_protein_count) / initial_protein_count * 100:.2f}% of unique proteins.")

# Now filter the full dataset to keep only rows with valid protein IDs.
valid_protein_ids = set(valid_protein_df["BindingDB MonomerID"])
filtered_data = filtered_data[filtered_data["BindingDB MonomerID"].isin(valid_protein_ids)]

# 3.5. Create a new mapping for protein groups (to avoid issues with commas)
# We'll assign new identifiers like "P_1", "P_2", ...
unique_protein_ids = sorted(valid_protein_ids)
protein_mapping = {orig: f"P_{i}" for i, orig in enumerate(unique_protein_ids, start=1)}

# 6. Glycan Filtering: Filter out problematic glycan SMILES using RDKit fingerprint generation.
ligand_groups = filtered_data.dropna(subset=["BindingDB Ligand Name"]).drop_duplicates(subset=["BindingDB Ligand Name"])
initial_ligand_count = len(ligand_groups)
valid_ligand_rows = []

# Pre-create the fingerprint generator.
fingerprint_generator = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048, includeChirality=True)

for idx, row in ligand_groups.iterrows():
    smiles = row["Ligand SMILES"]
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            print(f"Skipping glycan at row {idx}: Unable to parse SMILES.")
            continue
        _ = fingerprint_generator.GetFingerprint(mol)
        valid_ligand_rows.append(row)
    except Exception as e:
        print(f"Skipping glycan at row {idx} due to error: {e}")
        continue

valid_ligand_df = pd.DataFrame(valid_ligand_rows)
filtered_ligand_count = len(valid_ligand_df)
print(f"Glycan Filtering: Removed {(initial_ligand_count - filtered_ligand_count) / initial_ligand_count * 100:.2f}% of unique glycans.")

# Build glycan mapping: assign new names like "G_1", "G_2", ...
valid_glycan_set = set(valid_ligand_df["BindingDB Ligand Name"])
glycan_mapping = {orig: f"G_{i}" for i, orig in enumerate(sorted(valid_glycan_set), start=1)}

# Create the Glycan-Structures file using the new glycan names.
glycan_struct = pd.DataFrame({
    "Name": valid_ligand_df["BindingDB Ligand Name"].map(glycan_mapping),
    "IUPAC": "NA",  # No IUPAC string available
    "SMILES": valid_ligand_df["Ligand SMILES"]
})
glycan_struct.to_csv("BDB_Glycan-Structures-CFG611.txt", sep='\t', index=False)

# 4. Create Train_Fractions.csv
# Expand each protein-ligand pair into multiple rows spanning concentrations from 0.01 to 200.
num_points = 10  # Number of concentration points
concentration_values = np.linspace(0.01, 200, num_points)

train_rows = []
obj_id = 1
for _, row in filtered_data.iterrows():
    ic50 = row["IC50 (nM)"]
    # Replace the original protein id with the new sanitized version.
    protein_group_new = protein_mapping[row["BindingDB MonomerID"]]
    # Also, use the glycan mapping later; for now store the original glycan id.
    original_glycan = row["BindingDB Ligand Name"]
    for conc in concentration_values:
        f_value = conc / (ic50 + conc)
        train_rows.append({
            "ObjId": obj_id,
            "ProteinGroup": protein_group_new,
            "Concentration": conc,
            "GlycanID": original_glycan,  # will map to new name below
            "f": f_value
        })
        obj_id += 1

train_fractions = pd.DataFrame(train_rows)

# Additional filtering: Remove rows with GlycanID not present in the valid glycan set.
initial_fraction_count = len(train_fractions)
train_fractions = train_fractions[train_fractions["GlycanID"].isin(valid_glycan_set)]
filtered_fraction_count = len(train_fractions)
print(f"Train Fractions Filtering: Removed {(initial_fraction_count - filtered_fraction_count) / initial_fraction_count * 100:.2f}% of rows due to invalid GlycanIDs.")

# Map original glycan names to the new names in the Train_Fractions file.
train_fractions["GlycanID"] = train_fractions["GlycanID"].map(glycan_mapping)

train_fractions.to_csv("BDB_Train_Fractions.tsv", index=False, sep='\t')

# 5. Create Protein-Sequence-Table.txt using the valid proteins and new protein names.
protein_seq = pd.DataFrame({
    "ProteinGroup": valid_protein_df["BindingDB MonomerID"].map(protein_mapping),
    "Accession": valid_protein_df["BindingDB MonomerID"].astype(str),  # Using original monomer id as surrogate
    "Uniprot": valid_protein_df["Target Name"],
    "Description": ("Target: " + valid_protein_df["Target Name"].astype(str) +
                    " (Organism: " + valid_protein_df["Target Source Organism According to Curator or DataSource"].astype(str) + ")"),
    "Amino Acid Sequence": valid_protein_df[seq_column] if seq_column is not None else "NA"
})
protein_seq.to_csv("BDB_Protein-Sequence-Table.txt", sep='\t', index=False)

print("Conversion complete! All three files have been generated for your training pipeline.")
```
